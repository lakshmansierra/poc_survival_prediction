# Data Preprocessing Tools

## Importing the libraries

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
# Load Excel files
df = pd.read_csv("1_service_now_survival_data/1_service_now_survival_data.csv")
```

## Correlation

```python
print(df.isna().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Drop non-numeric and identifier columns (like 'Asset_Number', 'start', 'stop', 'event' can be kept for inspection if needed)
numeric_df = df.drop(columns=["Asset_Number", "Category"])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```
